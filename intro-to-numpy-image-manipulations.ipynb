[stackoverflow](https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python)

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

```python
img = plt.imread('./images/toy-car.jpg')
toy_car = rgb2gray(img)
```

```python
toy_car.shape
```

```python
plt.imshow(toy_car, cmap = plt.get_cmap('gray'))
```

```python
toy_car[250, 200]
```

```python
toy_car[200:400, 300:400]=0
plt.imshow(toy_car, cmap = plt.get_cmap('gray'))
```
